```python
cohort_but_not_in_2021_cross = []
more_than_twice = []
for p in repeat_participantsV2:
    p_df = original_df[original_df["UNIQUE_id"] == p].dropna(axis=1, how="all")
    inside_2021_cross = False
    inside_2022_cohort = False
    inside_2023_cohort = False
    count = 0
    for r in p_df["DATASET"]:
        count += 1
        if r == "2021 Cross-Sectional":
            inside_2021_cross = True
        if r in ["2022 Cohort", "2023 Cohort"]:
            if r == "2022 Cohort": inside_2022_cohort = True
            if r == "2023 Cohort": inside_2023_cohort = True

    if not inside_2021_cross and inside_2022_cohort and inside_2023_cohort:
        cohort_but_not_in_2021_cross.append(p)
        if count > 2:
            more_than_twice.append(p)
    
print(len(cohort_but_not_in_2021_cross), len(set(cohort_but_not_in_2021_cross)))
print(more_than_twice)
print(set(cohort_but_not_in_2021_cross))
```

```python
in_2022_cohort_and_cross = []
for p in set(original_df["UNIQUE_id"]):
    p_df = original_df[original_df["UNIQUE_id"] == p].dropna(axis=1, how="all")
    inside_2022_cohort = False
    inside_2022_cross = False
    for r in p_df["DATASET"]:
        if r == "2021 Cross-Sectional":
            inside_2022_cross = True
        if r == "2022 Cohort":
            inside_2022_cohort = True

    if inside_2022_cross and inside_2022_cohort:
        in_2022_cohort_and_cross.append(p)

print(len(in_2022_cohort_and_cross), len(set(in_2022_cohort_and_cross)))
print(in_2022_cohort_and_cross)
```
