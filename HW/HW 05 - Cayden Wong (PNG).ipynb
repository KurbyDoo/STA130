In statistics, the population parameter (denoted by $\mu$) is a value that describes the population, for example the average height of all people in Canada is a population parameter for the population Canada. It is a true unknown value that is impossible to calculate, however, we can estimate its value through guessing. When we have a guess about population parameter $\mu$, we perform a process known as hypothesis testing to determine whether it is likely for the guess to be false. More specifically, hypothesis testing is the process of collecting and analyzing data to determine whether there is sufficient evidence against this guess, called "the null hypothesis" $\mu_0$. Usually, we associate our guess, the null hypothesis, with the opposite to what we want to show, since it is easier to show something is not likely true than show something is likely to be true. For example, if I want to show that effect of a new medical drug impacts a patients medical heath, it would be easier to let our null hypothesis be that the drug has no effect and then use statistical evidence to show why it is unlikely that the drug has no effect. To perform hypothesis testing on the null hypothesis, we can perform tests on a sample (a subset of the population) instead of the entire population. From this sample, we can collect information about the sample and its sample statistics such as the mean, denoted by $\bar x$. Then, while performing hypothesis testing, we generate many different sample statistics, denoted by $x_i$ and compare them to our initial sample statistic. Using the results of the comparisons, we can make conclusions about the population and where the population parameter $\mu$ may lie without needing to collect data from the entire population and otherwise impossible task. Our initial goal was to determine information about the population, since our hypothesis tests make comparisons between individual sample statistics $x_i$ and our initial sample statistic $\bar x$ to reach a conclusion, we are not making any conclusions about the initial sample statistic as it is just a value derived from a sample of our population, the group we want to make conclusions about. Instead we are using $\bar x$ to make conclusions about $\mu$ by showing whether there is statistical evidence against the likelihood of $\mu_0$ being true.

##### Hypothesis Testing: The Goal is to Disprove, Not Prove

- **The central idea of hypothesis testing is to assess the strength of evidence *against* a specific claim called the null hypothesis (H<sub>0</sub>).**  We never aim to "prove" the null hypothesis or any alternative hypothesis. Instead, we evaluate whether the data provide sufficient evidence to *reject* the null hypothesis. Our decisions are about rejecting or failing to reject, not about proving something true or false.
- **The null hypothesis is often a statement of "no effect" or "no difference."** For example, in a drug trial, the null hypothesis might be that the drug has no effect on a particular health outcome.

##### The Process of Hypothesis Testing

1.  **Define the null hypothesis (H<sub>0</sub>) and the alternative hypothesis (H<sub>A</sub>).** The null hypothesis is the claim we are trying to find evidence against, while the alternative hypothesis encompasses other possibilities. 
2.  **Assume the null hypothesis is true and simulate samples under this assumption.** This creates what's called the *sampling distribution under the null hypothesis.* This distribution shows us the range of values we'd expect to see for our sample statistic (like the sample mean) if the null hypothesis were true and we repeatedly drew samples. 
3.  **Compare our observed sample statistic to the sampling distribution under the null hypothesis.** If the observed statistic falls in an extreme tail of this distribution  meaning it's quite unlikely to have occurred if H<sub>0</sub> were true  this provides evidence *against* the null hypothesis.
4.  **Quantify the strength of the evidence against H<sub>0</sub>**. We typically use a p-value to measure how likely it is to observe data as extreme as ours (or even more extreme) if the null hypothesis were true. A small p-value indicates stronger evidence against the null hypothesis.
5.  **Make a decision about whether to reject the null hypothesis based on the strength of the evidence.** This decision often involves comparing the p-value to a pre-determined significance level (), commonly set at 0.05. 

##### Important Distinctions and Considerations

- **Confidence levels are associated with confidence intervals, *not* hypothesis tests.** While confidence intervals can be used to perform hypothesis tests, it's essential to understand the difference between a confidence level (a property of the confidence interval procedure) and the strength of evidence against the null hypothesis (assessed in a hypothesis test).
- **We often simulate samples from a sample of the original population, a technique called *bootstrapping*.** This is especially relevant when we don't have access to the entire population, which is common in real-world data analysis.
- **Sample size (n) plays a crucial role in hypothesis testing.** The larger the sample size, the less variability we expect in our sample statistic and the more confident we can be in our conclusions.
- **The sources you provided emphasize the importance of using clear and precise language when discussing hypothesis testing.**  Avoid using terms like "prove" or "true" when describing the outcome of a hypothesis test. Instead, focus on the strength of evidence against the null hypothesis and whether the data support rejecting it.

Explain why this is in your own words in a way that makes the most sense to you, clarifying the meaning of any technical statistical terminology you use in your answer.

The p-value is defined as the probability where, under the conditions where the null hypothesis is true, the given or observed data occurred. A larger p-value would represent a larger probability of our observed data occurring, while a smaller p-value would represent a lower probability of our observed data occurring. If we assume that the null hypothesis is true (the hypothesis that there is no effect / no difference), and calculate an extremely low p-value, then we would determine that the probability of our observed event occurring is almost impossible, causing the null hypothesis to lose credibility. Typically, if we observe a p-value of less than 0.001 (or 1/1000 chance of observing a statistic as or more extreme than our observed statistic) then we can say there is very strong evidence against the null hypothesis as it is very unlikely for such an event to occur under the assumption that the null hypothesis is true. We would then reject the null hypothesis, claiming it was true would imply we observed a very unlikely observed sample statistic and are overlooking other potential factors that may have led to this result.

We want to determine whether humans have a tendency to tilt in one direction over the other when kissing. Lets say the null hypothesis is that humans do not have a preference in head tilt direction when kissing, i.e. the they will equally tilt their head in both directions (50/50 chance for left or right). Assuming the null hypothesis is true, let us determine the likely hood of observing 80/124 humans having a head tilt towards the right when kissing.

```python
import numpy as np
import plotly.express as ex

np.random.seed(0)

observed_statistic = 0.645
population_parameter_under_h0 = 0.5

sample_size = 124
number_of_simulations = 100000

simulated_statistics = np.zeros(number_of_simulations)

for i in range(number_of_simulations):
    random_head_tilts = np.random.choice([0, 1], size=sample_size)
    simulated_statistics[i] = random_head_tilts.mean()   

double_ended_more_extreme = abs(simulated_statistics - population_parameter_under_h0) >= abs(observed_statistic - population_parameter_under_h0)
```

```python
import plotly.graph_objs as go
import plotly.figure_factory as ff
# figure for demonstration only: code details not of primary concern

hist_data = [simulated_statistics+np.random.uniform(-0.05,0.05,size=len(simulated_statistics))]
group_labels = ['Bootstrap<br>Sampling<br>Distribution<br>of the<br>Sample<br>Mean<br><br>assuming<br>that the<br>H0 null<br>hypothesis<br>IS TRUE']
fig = ff.create_distplot(hist_data, group_labels, curve_type='normal',
                         show_hist=True, show_rug=False, bin_size=0.01)
pv_y = 8
pv_y_ = .50
fig.add_shape(type="line", x0=observed_statistic, y0=0, 
              x1=observed_statistic, y1=pv_y,
              line=dict(color="Green", width=4), name="Observed Test Statistic")
fig.add_trace(go.Scatter(x=[observed_statistic], y=[pv_y+pv_y_], 
                         text=["Observed<br>Test Statistic<br>^"], mode="text", showlegend=False))
# "as or more extreme" also include the "symmetric" observed test statistic...
symmetric_test_statistic = population_parameter_under_h0 -\
                           abs(observed_statistic-population_parameter_under_h0)
fig.add_shape(type="line", x0=symmetric_test_statistic, y0=0, 
              x1=symmetric_test_statistic, y1=pv_y,
              line=dict(color="Green", width=4), name="Observed Test Statistic")
fig.add_trace(go.Scatter(x=[symmetric_test_statistic], y=[pv_y+pv_y_], 
                         text=['"Symmetric" Observed Test Statistic<br>addrdssing for "as or more extreme"<br>^'], mode="text", showlegend=False))

# Add a transparent rectangle for the lower extreme region
fig.add_shape(type="rect", x0=0.25, y0=0, x1=symmetric_test_statistic, y1=pv_y,
              fillcolor="LightCoral", opacity=0.5, line_width=0)
# Add a transparent rectangle for the upper extreme region
fig.add_shape(type="rect", x0=observed_statistic, y0=0, x1=0.75, y1=pv_y,
              fillcolor="LightCoral", opacity=0.5, line_width=0)

# Update layout
fig.update_layout(
    title="Bootstrapped Sampling Distribution<br>under H0 with p-value regions",
    xaxis_title="Mean Health Score Change", yaxis_title="Density", yaxis=dict(range=[0, pv_y+2*pv_y_]))
fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
```

```python
p_value = double_ended_more_extreme.sum() / number_of_simulations
print("Number of Simulations: ", number_of_simulations, "\n\n",
      "Number of simulated test statistics (under HO)\n",
      'that are "as or more extreme" than the observed test statistic: ',
      double_ended_more_extreme.sum(), "\n\n",
      'p-value\n(= simulations "as or more extreme" / total simulations): ', p_value, sep="")
```

### The problem
Given a cup of tea and milk, We want to determine whether the average STA130 student can correctly determine what order the milk and tea were poured into the cup. 
If they are able to correctly identify the order in which they were poured, then this must mean that there exists a difference in taste based on the pouring order.
Is Dr.Bristol correct in saying that the tea tastes different when milk is poured before or after the tea? Or is there no difference in taste and she simply got lucky.

Here the population parameter is, the probability that the average STA130 student can correctly guess the correct pouring order of milk and tea.  
We state that the null hypothesis ($H_0$) is that there is no difference between the taste between tea in which milk was poured first or after. That is formally, that the population parameter is 0.5.  
The alternative hypothesis ($H_A$) is that there exists a difference in taste depending on pouring order. That is formally, the population parameter $\mu$ is not 0.5.

We have run an experiment on a sample of 80 STA130 students to determine how many of them can correctly identify the pouring order of the tea. We found that 49/80 total students were able to correctly identify the pouring order.

```python
tea_observed_correctly_guessed = 49
tea_observed_sample_size = 80
tea_observed_statistic = tea_observed_correctly_guessed / tea_observed_sample_size
print(f"Our observations show that {(tea_observed_statistic * 100):.2f}% of the sample students correctly identified the pouring order.")
```

How likely is it for this to happen if we assumed the null hypothesis was true? Is 61.25% correct guess rate a significant amount with a random sample of 80 students from the population?

Lets do a two sided hypothesis test under the assumption the null hypothesis is true, and compare the null distribution with our observed sample statistic. 


```python
import numpy as np
np.random.seed(7)

tea_pop_param_value_under_h0 = 0.5

tea_number_of_simulations = 10000

tea_sample_means = np.zeros(tea_number_of_simulations)
for i in range(tea_number_of_simulations):
    # Simulating under H_0, that is there is a 50% chance that a student guesses correctly
    tea_sample = np.random.choice([0, 1], size=tea_observed_sample_size, replace=True)
    tea_sample_means[i] = tea_sample.mean()
    
```

```python
import plotly.graph_objects as go
import plotly.figure_factory as ff

tea_histogram_data = [tea_sample_means + np.random.uniform(-0.05,0.05,size=len(tea_sample_means))] # Add random variation for a smoother histogram

group_labels = ['Bootstrap<br>Sampling<br>Distribution<br>of the<br>Sample<br>Mean<br><br>assuming<br>that the<br>H0 null<br>hypothesis<br>IS TRUE']
fig = ff.create_distplot(tea_histogram_data, group_labels, curve_type='normal',
                         show_hist=True, show_rug=False, bin_size=0.01)
pv_y = 8
pv_y_ = .50
fig.add_shape(type="line", x0=tea_observed_statistic, y0=0, 
              x1=tea_observed_statistic, y1=pv_y,
              line=dict(color="Green", width=4), name="Observed Test Statistic")

fig.add_trace(go.Scatter(x=[tea_observed_statistic], y=[pv_y+pv_y_], 
                         text=["Observed<br>Test Statistic<br>^"], mode="text", showlegend=False))

tea_symmetric_test_statistic = tea_pop_param_value_under_h0 -\
                           abs(tea_observed_statistic - tea_pop_param_value_under_h0)

fig.add_shape(type="line", x0=tea_symmetric_test_statistic, y0=0, 
              x1=tea_symmetric_test_statistic, y1=pv_y,
              line=dict(color="Green", width=4), name="Observed Test Statistic")

fig.add_trace(go.Scatter(x=[tea_symmetric_test_statistic], y=[pv_y+pv_y_], 
                         text=['"Symmetric" Observed Test Statistic<br>addressing for "as or more extreme"<br>^'], mode="text", showlegend=False))

# Add a transparent rectangle for the lower extreme region
fig.add_shape(type="rect", x0=0.25, y0=0, x1=tea_symmetric_test_statistic, y1=pv_y,
              fillcolor="LightCoral", opacity=0.5, line_width=0)
# Add a transparent rectangle for the upper extreme region
fig.add_shape(type="rect", x0=tea_observed_statistic, y0=0, x1=0.75, y1=pv_y,
              fillcolor="LightCoral", opacity=0.5, line_width=0)

# Update layout
fig.update_layout(
    title="Bootstrapped Sampling Distribution<br>under H0 with p-value regions",
    xaxis_title="Mean Correct Guesses",
    yaxis_title="Density",
    yaxis=dict(range=[0, pv_y+2*pv_y_]),
    height=700)
fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
```

Above we see the distribution of correct guesses under the assumption that the null hypothesis is true. We can observe that the red shaded area under the graph corresponds to the likelihood of encountering a test statistic as or more extreme to our observed test statistic, also known as the p-value. What is the area of this region? How small is it compared to the total area? Is the p-value significantly small enough to correspond with strong evidence against the null hypothesis?

Lets calculate this value and see if it falls below $\alpha = 0.05$, or falls outside an interval that covers 95% of test statistics under the assumption that the null hypothesis is true. If our calculated p-value is below $\alpha$, we say with 95% confidence that there is sufficient evidence to reject the null hypothesis.

```python
tolerance = 1e-6

# Calculate the same as above with precision fixed
tea_sim_stats_as_or_more_extreme_than_obs_stat = np.logical_or(
    abs(tea_sample_means - tea_pop_param_value_under_h0) >= abs(tea_observed_statistic - tea_pop_param_value_under_h0),
    np.isclose(abs(tea_sample_means - tea_pop_param_value_under_h0), abs(tea_observed_statistic - tea_pop_param_value_under_h0), atol=tolerance)
)

tea_p_value_as_or_more_extreme = tea_sim_stats_as_or_more_extreme_than_obs_stat.sum() / tea_number_of_simulations
print(f"Number of Simulations: {tea_number_of_simulations}\n\n",
      "Number of simulated test statistics (under HO)\n",
      f'that are "as or more extreme" than the observed test statistic: {tea_sim_stats_as_or_more_extreme_than_obs_stat.sum()}\n',
      f'P_VALUE: (= simulations "as or more extreme" / total simulations): {tea_p_value_as_or_more_extreme}\n', sep="")
```

After running 10000 simulations of samples under the null hypothesis, we have found that approximately 5.1% of simulated test statistics are as or more extreme than our observed test statistic. This is almost enough to say with 5% significance level that there is sufficient evidence against the null hypothesis.

What if we change our method of approaching the problem? We want to know whether it is MORE likely for students to guess the correct pouring order than at complete random.
Instead, lets change our alternative hypothesis to consider whether the chance of guessing correctly is greater than complete random. Now when calculating our p-value, we are only going to record test statistics that were greater than or observed statistic. 
That is formally, that the population parameter lies above 0.5. To compensate, our null hypothesis will now be whether the population parameter is equal to or less than 0.5.

```python
# Calculate more extreme with precision
tea_sim_stats_more_extreme_than_obs_stat = np.logical_or(
    tea_sample_means >= tea_observed_statistic,
    np.isclose(tea_sample_means, tea_observed_statistic, atol=tolerance)
)


tea_p_value_more_extreme = tea_sim_stats_more_extreme_than_obs_stat.sum() / tea_number_of_simulations
print(f"Number of Simulations: {tea_number_of_simulations}\n\n",
      "Number of simulated test statistics (under HO)\n",
      f'that are "more extreme" than the observed test statistic: {tea_sim_stats_more_extreme_than_obs_stat.sum()}\n\n',
      f'P_VALUE: = simulations "more extreme" (>= {tea_observed_statistic:.4f}) / total simulations): {tea_p_value_more_extreme}\n', sep="")
```

Now, we find that approximately 2.7% of simulated test statistics as or more extreme than our observed statistic in the positive direction. This indicates strong evidence against the null hypothesis which implies strong evidence to support the alternative hypothesis with a 5% significance level or chance of Type I error.

Initially, using two sided hypothesis testing with p-values did not provide sufficient evidence to reject the null hypothesis thus forcing us to chance our alternative hypothesis to a more specific hypothesis. Afterwards, we were able to calculate a p-value that fell below a 5% significance level, indicating sufficient evidence against the null hypothesis.

What if we tried to reject the null hypothesis instead using a confidence interval over p-values? Let's perform bootstrapping on our observed sample and determine a 95% confidence interval. Will we find a different conclusion?

```python
import numpy as np
np.random.seed(7)


tea_bootstrapped_number_of_simulations = 10000
tea_bootstrapped_sample_means = np.zeros(tea_bootstrapped_number_of_simulations)

for i in range(tea_bootstrapped_number_of_simulations):
    tea_sample = np.random.choice([0, 1], p=[(tea_observed_sample_size - tea_observed_correctly_guessed)/tea_observed_sample_size, tea_observed_correctly_guessed/tea_observed_sample_size], size=tea_observed_sample_size, replace=True)
    tea_bootstrapped_sample_means[i] = tea_sample.mean()

# Get unique values and their counts
unique_values, counts = np.unique(tea_bootstrapped_sample_means, return_counts=True)

# Get the number of unique values
number_of_unique_values = len(unique_values)

# Display the results
print("Unique values:", unique_values)
print("Counts of each unique value:", counts)
print("Number of unique values:", number_of_unique_values)

print(f"Mean of distribution is {tea_bootstrapped_sample_means.mean()}")

    
```

```python
# Calculate 95% confidence interval for the bootstrapped sample means
conf_int_low_bootstrap = np.percentile(tea_bootstrapped_sample_means, 2.5)
conf_int_high_bootstrap = np.percentile(tea_bootstrapped_sample_means, 97.5)

# Create a figure factory plot (distribution plot)
group_labels = ['Bootstrapped\nSampling\nDistribution\nof the\nSample\nMean']

tea_bootstrapped_histogram_data = tea_bootstrapped_sample_means + np.random.uniform(-0.05,0.05,size=len(tea_bootstrapped_sample_means))
# tea_bootstrapped_histogram_data = [tea_bootstrapped_sample_means]

fig = ff.create_distplot(
    [tea_bootstrapped_histogram_data],
    group_labels,
    show_hist=True,
    show_rug=False,
    curve_type='normal',
    bin_size=0.01
)

# Add vertical lines for confidence interval
fig.add_shape(
    type="line",
    x0=conf_int_low_bootstrap,
    y0=0,
    x1=conf_int_low_bootstrap,
    y1=8,  # Adjust based on the y-axis range
    line=dict(color="red", width=2, dash="dash"),
    name="2.5% Percentile (Lower CI)"
)

fig.add_shape(
    type="line",
    x0=conf_int_high_bootstrap,
    y0=0,
    x1=conf_int_high_bootstrap,
    y1=8,  # Adjust based on the y-axis range
    line=dict(color="green", width=2, dash="dash"),
    name="97.5% Percentile (Upper CI)"
)

# Update layout
fig.update_layout(
    title="Bootstrapped Sampling Distribution\nwith 95% Confidence Interval",
    xaxis_title="Mean Correct Guesses",
    yaxis_title="Density",
    showlegend=False,
    height=700
)

fig.show(renderer="png")  # Display the plot
```

```python
print(f"The bootstrapped confidence interval covers values between {conf_int_low_bootstrap} - {conf_int_high_bootstrap}")
```

Above, we found that the confidence interval spans between 0.5 - 0.7125 which lies directly on the null hypothesis (the probability of guessing correctly being 0.5). Again, with a symmetric two sided test, we fail to reject the null hypothesis and cannot say with 95% confidence that the null hypothesis is false. Can we perform the same narrowing of the alternative hypothesis and now find an interval that give us 95% confidence that the correct guess occurs more frequently then at random?

Applying the same concept of one-sided versus two-sided tests, we can instead consider the confidence interval that spans the upper 95% of the data.

```python
# Calculate 95% confidence interval for the bootstrapped sample means
conf_int_low_bootstrap_top_95 = np.percentile(tea_bootstrapped_sample_means, 5)

# Create a figure factory plot (distribution plot)
group_labels = ['Bootstrapped\nSampling\nDistribution\nof the\nSample\nMean']
tea_bootstrapped_histogram_data = tea_bootstrapped_sample_means + np.random.uniform(-0.05,0.05,size=len(tea_bootstrapped_sample_means))

# tea_bootstrapped_histogram_data = [tea_bootstrapped_sample_means]

fig = ff.create_distplot(
    [tea_bootstrapped_histogram_data],
    group_labels,
    show_hist=True,
    show_rug=False,
    curve_type='normal',
    bin_size=0.01
)

# Add vertical lines for confidence interval
fig.add_shape(
    type="line",
    x0=conf_int_low_bootstrap_top_95,
    y0=0,
    x1=conf_int_low_bootstrap_top_95,
    y1=8,  # Adjust based on the y-axis range
    line=dict(color="red", width=2, dash="dash"),
    name="5% Percentile (Lower CI)"
)

# Update layout
fig.update_layout(
    title="Bootstrapped Sampling Distribution\nwith 95% Confidence Interval",
    xaxis_title="Mean Correct Guesses",
    yaxis_title="Density",
    showlegend=False,
    height=700
)

fig.show(renderer="png")  # Display the plot
```

```python
print(f"The bootstrapped confidence interval covers values between {conf_int_low_bootstrap_top_95} - {np.inf}")
```

Again, after narrowing our alternative hypothesis, our new one-sided confidence interval now provides sufficient information to reject the null hypothesis with 95% confidence.

### Conclusion
Using both methods of hypothesis testing, confidence intervals and p-value hypothesis testing, there is not sufficient evidence to reject the null hypothesis. However, when we narrow our alternative hypothesis to focus on determining whether the population parameter lies above 0.5 (the chance of guessing correctly randomly), we can conclude that there is strong evidence to reject the null hypothesis with a 5% significance level from our one-sided p-value hypothesis test and that there is sufficient evidence against the null hypothesis since 0.5 lies outside of our one-sided bootstrapped confidence interval.

Thus our final conclusion is that there is not enough statistical evidence to reject whether the chance that a student correctly is completely random, but there is significant evidence against the the chance the students guess is less than or equal to random chance. That is, there is sufficient evidence to support the alternative hypothesis that the probability of a student correctly guessing the correct pouring order is greater than random chance, i.e. a 50% chance of guessing correctly.

We can infer that there is likely a difference in taste between cups of tea with milk poured before versus milk poured after the tea, allowing the average STA130 student to differentiate and correctly guess the pouring order of the milk and tea.
