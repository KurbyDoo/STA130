**Why is the number of variables different that what is described in `.shape` using `.describe()`?**
When using `.describe()` without any additional parameters, the function only returns data about variables that contain only integer values. This means columns that include information such as name, class, and embark_town are not included when using `.describe()` leading to a lower percived number of variables. This can be fixed to include all types by adding the following additional argument: `.describe(include='all')`.

**Why is the count number different between variables?**
The count number is different to the number described by `.shape` as it does not count missing values contained within the dataset. For example, the variable survived returns  a count of `891` (the same as described by `.shape`) while age only returns a count of `714`. This likely indicates that there is missing data about the age of some of the observations within the dataset, most likely information that was unobtainable because the passenger may have not survived.

**Attribute** (Example: `df.shape`)
An attribute is an intrinsic property of some object that stores information or data about that object (in the case of `.shape`, it is a property describing the dimensions of a pandas DataFrame object). It is not a function call and does not need to be called with any parameters.

**Method** (Example: `df.describe()`)
A method is an property of some object that performs and operation / returns information about said object (for .describe(), this is a method of the pandas DataFrame object). Since it is a function call, it can take in additional paramaters within the `()`.
