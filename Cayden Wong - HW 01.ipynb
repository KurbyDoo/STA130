### Terminology

**Count**: The number of null data points for the given variable.  
**Mean**: The average of all the data points ($\bar x = \frac{1}{n}\sum^n_{i=1}x_i$).  
**Standard Deviation (std)**: A figure representing the variance between the mean and the individual data points. Low values represent little varience, larger values represent larger varience.  
**Minimum (min)**: The smallest element within a given variable.  
**First Quartile (25%)**: The value which 25% of the data points fall below.  
**Median (50%)**: The value which 50% of the data points fall below.  
**Third Quartile (75%)**: The value which 75% of data points fall below.  
**Maximum (max)**: The argest element within a given variable.

##### Provide an example of a "use case" in which using `df.dropna()` might be peferred over using del `df['col']`

One might prefer to use `df.dropna()` over `df['col']` if their data set contains observations that are missing data-points for several different varaibles. Deleting these observations using `df.dropna()` may be best as their incompleteness may skew / portray the general data in an inaccurate manner. Attempting to remove the few missing data-points using `df['col']` would remove many data-points that do exists in most other observations.

##### Provide an example of "the opposite use case" in which using del `df['col']` might be preferred over using `df.dropna()`

`df['col']`s use may be preferred if many observations contain missing data-points for one specific variable. Since `df.dropna()` would remove all of the observations with missing data, many data points that would provide insight on other variables within the dataset would be lost. Contrarily, `df['col']` would discard the variable containg mostly missing information without losing data points about other variables from the rest of each observation.

##### Discuss why applying `del df['col']` before `df.dropna()` when both are used together could be important

Since `df.dropna()` removes all observations with missing data, if we are using `del df['col']` to remove a column that contains many points of missing data, we would want to use it before `df.dropna()`. Consider the scenario where many observations only contain a singular missing data-point all under a single variable. Using `df.dropna()` first would result in all of those observations being removed from the dataset. However, with use of `del df['col']`, we can remove the variable with many missing data-points before `df.dropna()` is run thus preserving the observations which would have otherwise been deleted.
